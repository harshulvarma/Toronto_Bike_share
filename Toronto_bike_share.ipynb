```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
dataQ1 = pd.read_csv('2017_BikeShare_Data/2017 Data/Bikeshare Ridership (2017 Q1).csv')
dataQ2 = pd.read_csv('2017_BikeShare_Data/2017 Data/Bikeshare Ridership (2017 Q2).csv')
dataQ3 = pd.read_csv('2017_BikeShare_Data/2017 Data/Bikeshare Ridership (2017 Q3).csv')
dataQ4 = pd.read_csv('2017_BikeShare_Data/2017 Data/Bikeshare Ridership (2017 Q4).csv')
```

```python
dataQ1.head()
```

```python
dataQ2.head()
```

```python
dataQ3.head()
```

```python
dataQ4.head()
```

```python
dataQ1 = dataQ1.drop(columns = ['from_station_id','to_station_id'])
dataQ2 = dataQ2.drop(columns = ['from_station_id','to_station_id'])
```

```python
dataQ1.insert(0, 'Quarter', 'Q1')
dataQ2.insert(0, 'Quarter', 'Q2')
dataQ3.insert(0, 'Quarter', 'Q3')
dataQ4.insert(0, 'Quarter', 'Q4')
```

```python
dataQ1.head()
```

```python
bikedata2017 = [dataQ1,dataQ2,dataQ3,dataQ4]
df = pd.concat(bikedata2017)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['trip_start_time'] = pd.to_datetime(df['trip_start_time'])
#df['trip_stop_time'] = pd.to_datetime(df['trip_stop_time'])
```

```python
df['date'] = df['trip_start_time'].dt.date
df['trip_start_at'] = df['trip_start_time'].dt.time
#df['trip_end_time'] = df['trip_stop_time'].dt.time
```

```python
df.head()
```

```python
countQ1_members = len(df[(df.Quarter == 'Q1') & (df.user_type == 'Member')])
countQ2_members = len(df[(df.Quarter == 'Q2') & (df.user_type == 'Member')])
countQ3_members = len(df[(df.Quarter == 'Q3') & (df.user_type == 'Member')])
countQ4_members = len(df[(df.Quarter == 'Q4') & (df.user_type == 'Member')])

countQ1_casuals = len(df[(df.Quarter == 'Q1') & (df.user_type == 'Casual')])
countQ2_casuals = len(df[(df.Quarter == 'Q2') & (df.user_type == 'Casual')])
countQ3_casuals = len(df[(df.Quarter == 'Q3') & (df.user_type == 'Casual')])
countQ4_casuals = len(df[(df.Quarter == 'Q4') & (df.user_type == 'Casual')])

trace1 = go.Bar( x=['Q1', 'Q2', 'Q3', 'Q4'],
               y=[countQ1_members,countQ2_members,countQ3_members,countQ4_members],
               name = 'Members',marker=dict(color='rgb(200,50,50)'))
trace2 = go.Bar( x=['Q1', 'Q2', 'Q3', 'Q4'],
               y=[countQ1_casuals,countQ2_casuals,countQ3_casuals,countQ4_casuals],
               name = 'Casuals',marker=dict(color='rgb(255,150,50)'))

data = [trace1, trace2]
layout = go.Layout(barmode='stack')
fig = go.Figure(data=data, layout=layout)
    
iplot(fig)
```

```python
df['day_of_week'] = df['trip_start_time'].dt.day_name()
df['trip_start_at_hr'] = df['trip_start_time'].dt.hour
```

```python
dftrips = pd.DataFrame(data = df, columns = ['trip_start_at_hr','day_of_week'])
```

```python
dftrips.head(10)
```

```python
dftrips = dftrips.groupby(['trip_start_at_hr','day_of_week']).day_of_week.agg('count').to_frame('z').reset_index()
```

```python
dftrips.head(8)
```

```python
data = [
    go.Heatmap(
        z=dftrips.z,
        x=dftrips.trip_start_at_hr,
        y=dftrips.day_of_week,
        colorscale=[[1.0, 'rgb(0,150,200)'], [0.9, 'rgb(0,100,200)'], [0.8, 'rgb(0,50,200)'], [0.7, 'rgb(0,0,200)'], [0.6, 'rgb(50,0,200)'], [0.5, 'rgb(100,0,200)'], [0.4, 'rgb(150,0,200)'], [0.3, 'rgb(200,0,200)'], [0.2, 'rgb(200,0,150)'], [0.1, 'rgb(200,0,100)'], [0.0, 'rgb(200,0,50)']],
        opacity = 0.8
    )
]

layout = go.Layout(
    title='Number of trips per hour per day',
    xaxis = dict(title = 'Hour of day',ticks='', nticks=36),
    yaxis = dict(title = 'Day of Week',ticks='' )
)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
dftriptime = pd.DataFrame(data = df, columns = ['trip_duration_seconds'])
```

```python
times = {}
for t in df.trip_duration_seconds:
    if t not in times:
        times[t] = 1
    else:
        times[t] += 1
```

```python
dftriptime['trips'] = pd.DataFrame.from_dict(times, orient = "index")
```

```python
dftriptime = dftriptime[dftriptime['trip_duration_seconds'] < 1800]
```

```python
dftriptime = dftriptime.groupby(['trip_duration_seconds']).sum().reset_index()
```

```python
dftriptime = dftriptime.sort_values(by='trips')
```

```python
dftriptime = dftriptime.reset_index(drop = True)
```

```python
dftriptime
```

```python
data = [go.Bar(
            x=dftriptime['trip_duration_seconds'],
            y=dftriptime['trips'],
    marker=dict(color=dftriptime.trips,opacity = 0.8)
    )]

iplot(data, filename='basic-bar')
```
